```python
# Imports
import sqlite3
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime, date
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from subprocess import call
from IPython.display import Image
```

# <font color = 'black' style='font-size: 26px;'> Importando os Dados </font>  
<hr style = 'border: 1.5px solid black;'>

<p style='line-height: 2; text-align: justify; text-indent: 35px;'>Os dados abaixos foram retirados do banco de dados disponibilizado para a resolução do exercício. A biblioteca utilizada para efetuar essa conecção foi a sqlite3. Todas as tabelas abaixo estão em sua forma bruta, apresentando todos os campos originais do modelo relacional do banco. </p>

```python
def get_table(query, columns_df):
    
    # Conectando ao Banco de Dados
    con = sqlite3.connect('../Dados/desafio-tecnico.db')
    cur = con.cursor()
    
    # Executando a query
    cur.execute(query)
    
    # Obtendo todas as linhas relacionadas a consulta
    rows = cur.fetchall()
    
    # Finalizando a conexão com o banco de dados
    cur.close()
    
    table = pd.DataFrame(rows, columns=columns_df)    
    return table 
```

```python
# Informações das contas
table_accounts = get_table('SELECT * FROM accounts', 
                           ['id', 'account_number', 'birth','occupation', 'email', 'address_id', 'created_at'])

table_accounts.head(5)
```

```python
# Endereço do proprietário da conta
table_address = get_table('SELECT * FROM address', ['id', 'state', 'city', 'created_at'])
table_address.head(5)
```

```python
# Classificação da conta
table_levels = get_table('SELECT * FROM levels', ['id', 'account_number', 'levels', 'created_at'])
table_levels.head(5)
```

```python
# Boletos emitidos e seus respectivos status
table_charges = get_table('SELECT * FROM charges', ['id', 'account_number', 'status', 'value', 'created_at'])
table_charges.head(5)
```

```python
# Transações efetuadas
table_transactions = get_table('SELECT * FROM transactions', ['id', 'account_number', 'transaction_type_id', 'value', 'created_at'])
table_transactions.head(5)
```

```python
# Transações efetuadas
table_transaction_type = get_table('SELECT * FROM transaction_type', ['id', 'description', 'description_long'])
table_transaction_type.head(5)
```

# <font color = 'black' style='font-size: 26px;'> Pré-Processamento dos Dados </font>  
<hr style = 'border: 1.5px solid black;'>

<p style='line-height: 2; text-align: justify; text-indent: 35px;'>Nessa etapa, iremos realizar o pré-processamento dos dados, removendo os campos que sejam de pouca, ou nenhuma, importância para criação do modelo de classificação. Além disso, é necessário realizar a transformação de algumas colunas pois estão em um formato no qual o modelo não sabe utilizar no processo de aprendizado.</p>

## Removendo Colunas 

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> A princípio, iremos remover todas as colunas que não agregam conhecimento para o processo de aprendizado do modelo de machine learning. Tais colunas, como ID, e-mail, created_ad, não são dados importantes para determinar se uma conta é falsa ou não. Assim, deve-se removê-las.</p>

```python
# Removendo as colunas irrelevantes para o modelo
table_accounts.drop(['id', 'email'], axis=1, inplace=True)
table_accounts.head(5)
```

```python
# Removendo as colunas irrelevantes para o modelo
table_levels.drop(['id', 'created_at'], axis=1, inplace=True)
table_levels.head(5)
```

```python
# Removendo as colunas irrelevantes para o modelo
table_charges.drop(['id', 'value', 'created_at'], axis=1, inplace=True)
table_charges.head(5)
```

```python
# Removendo as colunas irrelevantes para o modelo
table_transactions.drop(['id', 'value', 'created_at'], axis=1, inplace=True)
table_transactions.head(5)
```

## Transformando os Dados

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Em seguida, é interessante manipular os campos birth e created_at da table_account a fim de obter informações como idade do titular e tempo de conta. Os algoritmos de machine learning não sabem manipular diretamente datas. Dessa forma, transformá-las em números inteiros é necessário para a sua utilização.  </p>

```python
# Transformando o tipo das colunas birth e created_at para datetime
table_accounts['birth'] = pd.to_datetime(table_accounts['birth']).dt.date
table_accounts['created_at'] = pd.to_datetime(table_accounts['created_at']).dt.date
```

```python
def age(born, div):    
    today = date.today()
    sub = today - born
    return sub.days // div

# Calculando a idade do titular
table_accounts['age_user'] = table_accounts['birth'].apply(age, div=365)

# Calculando a idade da conta
table_accounts['account_days'] = table_accounts['created_at'].apply(age, div=1)
```

```python
# Dropando as colunas
table_accounts.drop(['birth', 'created_at'], axis=1, inplace=True)
table_accounts.head(5)
```

## Boletos e Transações Efetuadas

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Certamente, o número de boletos (pagos e não pagos) juntamente com a quantidade de transações efetuadas podem ser informações valiosas para detecção de contas ilícitas. Dessa forma, deve-se extrair essas informações e, posteriormente, agrupá-las com os dados do titular. </p>

<p style='line-height: 2; text-align: justify; text-indent: 35px;'>   </p>

```python
def number_of_bank_slip(df, groupby_column):
    return df.groupby(by=groupby_column).count()

# Quantidade de boletos pagos
paid_bank_slips = number_of_bank_slip(table_charges[table_charges['status'] == 'paid'], 'account_number')
table_accounts = pd.merge(table_accounts, paid_bank_slips, how = 'inner', on = 'account_number')
table_accounts.rename(columns={'status' : 'paid_bank_slips'}, inplace=True)

# Quantidade de boletos não-pagos 
unpaid_bank_slips = number_of_bank_slip(table_charges[table_charges['status'] == 'unpaid'], 'account_number')
table_accounts = pd.merge(table_accounts, unpaid_bank_slips, how = 'inner', on = 'account_number')
table_accounts.rename(columns={'status' : 'unpaid_bank_slips'}, inplace=True)

# Total de boletos
table_accounts['total_bank_slips'] = table_accounts['paid_bank_slips'] + table_accounts['unpaid_bank_slips']
table_accounts.head(5)
```

```python
# Quantidade de transações do tipo 1
transaction_type_1 = number_of_bank_slip(table_transactions[table_transactions['transaction_type_id'] == 1], 'account_number')
table_accounts = pd.merge(table_accounts, transaction_type_1, how = 'inner', on = 'account_number')
table_accounts.rename(columns={'transaction_type_id' : 'received_bank_slips'}, inplace=True)

# Quantidade de transações do tipo 2
transaction_type_2 = number_of_bank_slip(table_transactions[table_transactions['transaction_type_id'] == 2], 'account_number')
table_accounts = pd.merge(table_accounts, transaction_type_2, how = 'inner', on = 'account_number')
table_accounts.rename(columns={'transaction_type_id' : 'pix_sent_external_account'}, inplace=True)

# Quantidade de transações do tipo 3
transaction_type_3 = number_of_bank_slip(table_transactions[table_transactions['transaction_type_id'] == 3], 'account_number')
table_accounts = pd.merge(table_accounts, transaction_type_3, how = 'inner', on = 'account_number')
table_accounts.rename(columns={'transaction_type_id' : 'pix_received'}, inplace=True)

# Total de transações
table_accounts['total_transaction'] = (table_accounts['received_bank_slips'] + table_accounts['pix_sent_external_account'] 
                                       + table_accounts['pix_received'])
table_accounts.head(5)
```

## Adicionando a Classificação da Conta

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Resumidamente, o desafio se trata de um problema de classificação supervisiono. A classificação da conta (A, B, C, D e F) são os rótulos dos dados de treinamento e teste. Dessa forma, deve-se dar um merge do número da conta com a sua respectiva classificação. </p>

```python
# Combinando a categorização da conta com as informações do titular
table_accounts = pd.merge(table_accounts, table_levels, on='account_number', how='inner')
table_accounts.head(5)
```

# <font color = 'black' style='font-size: 26px;'> Distribuição das Contas </font>  
<hr style = 'border: 1.5px solid black;'>

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> As classificações das contas estão distribuidas da forma abaixo. </p>

```python
# Palette and Style
sns.set_palette('viridis')
sns.set_style('darkgrid')

# Gráfico de barras
ax = sns.barplot(x=table_accounts.groupby('levels')['levels'].count(), y=['A','B','C','D','F'], orient='h')
ax.figure.set_size_inches(16,8)
ax.tick_params(labelsize=14)
ax.set_title('Distribuição das Contas', fontsize=18)
ax.set_xlabel("Quantidade", fontsize=16)
ax.set_ylabel("Classificação", fontsize=16)
ax
```

# <font color = 'black' style='font-size: 26px;'> Treino e Teste </font>  
<hr style = 'border: 1.5px solid black;'>

## Dividindo os dados 

```python
# Mapeando os labels
dict_labels = { 'A' : '0', 'B' : '0', 'C' : '0', 'D' : '0', 'F' : '1'}
table_accounts['levels'] = table_accounts['levels'].map(dict_labels)
```

```python
# Mapeando as ocupações em valores
dict_occupation = {'Outros' : 1, 'Corretor' : 2, 'Investidor' : 3, 'Desenvolvedor' : 4, 'Analista' : 5, 'Autonomo': 6}
table_accounts['occupation'] = table_accounts['occupation'].map(dict_occupation)
```

```python
# Variável Explicativas
X = table_accounts.drop(columns = ['levels']) 

# Variáveis Dependente
y = table_accounts['levels']
```

```python
# Dados de Treino e Teste
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=42, stratify=y)
```

## Algoritmo de Base (Baseline)

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Antes de criar o modelo de classificação, dé interessante estabelecer uma meta mínima a ser superada. Assim, um baseline é um modelo simples, que pode ser usado para mensurar o desempenho do modelo desenvolvido. </p>

```python
# Baseline algoritmo
baseline = DummyClassifier(strategy="most_frequent")
baseline.fit(X_train, y_train)
```

```python
# Predição
baseline.predict(X_test)

# Acurácia do Modelo
print(f'A acurácia do modelo baseline foi de {baseline.score(X_test, y_test).round(2)}')
```

## Random Forest

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Para o problema em questão, é interessante utilizar o algoritmo Random Forest. O Random Forest é um conjunto de árvores de decisões que podem ser utilizadas em problemas de classificação e regressão. Uma das principais vantagem desse modelo é que as escalas das features não interferem no processo de classificação, o que não ocorre no SVC por exemplo. Outra caracteristica interessante é que o próprio algoritmo sabe ponderar a importância de cada feature. </p>

```python
X_train.iloc[0]['occupation']
```

```python
# Treinando o modelo
model = RandomForestClassifier(max_depth=5)
model.fit(X_train, y_train)
```

```python
# Predições
y_pred = model.predict(X_test)

# Acurácia
print(f'A acurácia do modelo baseline foi de {model.score(X_test, y_test).round(2)}')
```

## Matriz de Confusão

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Analisando a matriz de confusão dos valores preditos, observa-se que o algoritmo classifica corretamente a maiora das instâncias. A taxa de acerto dos Verdadeiros Positivos (TP) foi de 99.5% (74,68% / 75%). De forma análoga, a taxa de acerto dos Verdadeiros Negativos foi de 76% (19.08% / 25%). Esse valor pior provavelmente se deve ao menor número de instâncias falsas (contas ilícitas) nos dados de treino.  </p>

```python
# Gerando a matriz de confusão
cf_matrix = confusion_matrix(y_test, y_pred)
sns.set(font_scale = 1.4)

# Plotando a matriz de confusão
ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, fmt='.2%', cmap='mako')
ax.figure.set_size_inches(8, 6)
ax.set_title('Matriz de Confusão', fontsize=20)
ax.set_xlabel('Valor Predito', fontsize=16)
ax.set_ylabel('Valor Real', fontsize=16)


ax.tick_params(labelsize=18)
```

## Features de Maior Importância

<p style='line-height: 2; text-align: justify; text-indent: 35px;'> Abaixo é apresentado o grau de importância atribuido a cada feature pelo modelo. Como era de se esperar, as features relacionadas as transações bancárias e emissões de boletos foram as mais importantes no processo de detecção de contas ilícitas.  </p>

```python
importances = pd.DataFrame()
importances['feature'] = X.columns
importances['importance'] = model.feature_importances_

importances.sort_values(by='importance', ascending=False, inplace=True)
```

```python
# Gráfico de barras
ax = sns.barplot(x=importances['importance'], y=importances['feature'], orient='h')
ax.figure.set_size_inches(16,8)
ax.tick_params(labelsize=14)
ax.set_title('Features de Maior Importância', fontsize=22)
ax.set_xlabel("% de Importância", fontsize=16)
ax.set_ylabel("Feature", fontsize=16)
ax
```
